```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load Titanic Dataset
import pandas as pd
titanic_df = pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")
```

```python
# Step 4: Bar plot for survival by gender
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(data=titanic_df, x='Survived', hue='Sex')
plt.title("Survival Count by Gender")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
#Assignment Task 3: Create 'Number of Relatives' column
titanic_df['Number of Relatives'] = titanic_df['SibSp'] + titanic_df['Parch']

# Scatter plot of passengers who were dead
dead_passengers = titanic_df[titanic_df['Survived'] == 0]
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dead_passengers, x='Number of Relatives', y='Fare', color='blue')
plt.title("Scatter Plot of Deceased Passengers: Number of Relatives vs Fare")
plt.xlabel("Number of Relatives")
plt.ylabel("Fare")
plt.show()
```
